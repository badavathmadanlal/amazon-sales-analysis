Monthly Profit

```python
df['Month'] = pd.to_datetime(df['Order Date']).dt.to_period('M')
```

```python
monthly_profit = df.groupby('Month')['Profit'].sum().reset_index()
```

```python
monthly_profit['Month'] = monthly_profit['Month'].astype(str)
```

```python
mean_profit = monthly_profit['Profit'].mean()
```

```python
plt.figure(figsize=(14,6))
mean = monthly_profit['Profit'].mean()
std = monthly_profit['Profit'].std()

colors = []

for p in monthly_profit['Profit']:
    if p > mean + std/2:
        colors.append('green')  #for high profit
    elif p < mean - std/2:
        colors.append('red') #for low profit
    else:
        colors.append('yellow') #for avg profit

plt.bar(monthly_profit['Month'], monthly_profit['Profit'], color=colors)

plt.xticks(rotation=90)

plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.title('Monthly Profit Trend')

plt.tight_layout()
plt.show()
```

**Amazon Sales & Profit Analysis Dashboard using Python**

```python
import matplotlib.pyplot as plt
import pandas as pd


df['Month'] = pd.to_datetime(df['Order Date']).dt.to_period('M').astype(str)
city_profit = df.groupby('City')['Profit'].sum().sort_values(ascending=False).head(10)
monthly_profit = df.groupby('Month')['Profit'].sum().reset_index()
category_profit = df.groupby('Category')['Profit'].sum()

plt.figure(figsize=(14,10))
plt.suptitle('Amazon Sales & Profit Analysis Dashboard using Python',
             fontsize=18, fontweight='bold')

#1:Top Cities
plt.subplot(2,2,1)
city_profit = city_profit.sort_values()
plt.barh(city_profit.index, city_profit.values, color='skyblue')
plt.title('Top Cities by Profit')
plt.xlabel('Profit')

#2:Monthly Trend
plt.subplot(2,2,2)
plt.plot(monthly_profit['Month'], monthly_profit['Profit'], marker='o',color='green')
plt.xticks(rotation=90)
plt.title('Monthly Profit Trend')
plt.xlabel('Month')
plt.ylabel('Profit')

# 3: Donut Chart
plt.subplot(2,2,3)
wedges, texts, autotexts = plt.pie(category_profit,
                                   labels=category_profit.index,
                                   autopct='%1.1f%%',
                                   startangle=140)
centre_circle = plt.Circle((0,0),0.60,fc='white')
plt.gca().add_artist(centre_circle)
plt.title('Profit Share by Category')

#4:Box Plot
plt.subplot(2,2,4)
plt.boxplot(df['Profit'])
plt.title('Profit Distribution')
plt.ylabel('Profit')

plt.tight_layout()
plt.show()
```

**City vs Category → Profit Heatmap**

```python
top_cities = df.groupby('City')['Profit'].sum().nlargest(10).index

filtered_df = df[df['City'].isin(top_cities)]
```

```python
pivot_table = filtered_df.pivot_table(
    values='Profit',
    index='City',
    columns='Category',
    aggfunc='sum'
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='coolwarm')

plt.title('Top 10 Cities vs Category Profit Heatmap')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df['Sales'],
    df['Profit'],
    s=df['Quantity']*10,
    c=df['Quantity'],
    cmap='viridis',
    alpha=0.6,
    label='Data Points'
)

plt.colorbar(label='Quantity')

plt.xlabel('Sales')
plt.ylabel('Profit')
plt.legend()
plt.title('Sales vs Profit Analysis (Bubble Size = Quantity)')

plt.show()
```
